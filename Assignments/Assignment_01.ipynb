Task 1: Create a pandas dataframe (dataframe name as 'df') with numpy random values (4 features and 4 observations)

```python
import pandas as pd
import numpy as np
```

```python
s1=pd.Series(np.random.rand(4))
s2=pd.Series(np.random.rand(4))
s3=pd.Series(np.random.rand(4))
s4=pd.Series(np.random.rand(4))
```

```python
df=pd.DataFrame([s1,s2,s3,s4])
df
```

Task 5: Display the 'Random value 2' and 'Random value 3' columns with location method and index location method

```python
#Column 2
loc = df.loc[:, 'Random value 2']
idx = df.iloc[:, 1]

print("\nRandom value 2 (Location method):\n", loc)
print("\nRandom value 2 (Index location method):\n", idx)
```

```python
#Column 3
loc = df.loc[:, 'Random value 3']
idx = df.iloc[:, 2]

print("\nRandom value 3 (Location method):\n", loc)
print("\nRandom value 3 (Index location method):\n", idx)
```
